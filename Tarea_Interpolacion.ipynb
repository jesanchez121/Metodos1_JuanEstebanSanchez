d. (Python) Para $f(x) = \sqrt{tan(x)}$ estimar la derivada central de orden $O(h^{2})$ en el
intervalo $[0.1, 1.1]$ con $h = 0.01$.


```python
def CentralDerivative(f,x,h):
  d = 0.
  if h != 0:
    d = (f(x+h) - f(x-h))/(2*h)
  return d

y2=CentralDerivative(f,x,h)
print(y2)
plt.plot(x,y2)
plt.show()
```

2. Compruebe que las funciones cardinales son base (i.e, $L_{i}(x) = δ_{ij}$ para
cada $j ∈ {0, 1, ..., n}$).
\
sea
$\delta_{ij} = \begin{cases}
1 & \text{si } i = j \\
0 & \text{si } i \neq j
\end{cases}$
\
Ya que $L_{i}(x) =  \prod_{j = 0 \ i \neq j}^n\frac{x − x_{j}}{x_{i} − x_{j}}$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
import pandas as pd

X = np.array([-3,-2,-1,0,1,2,3])
def Lagrange(x,X,i):

  L = 1
  for j in range(len(X)):
    if i != j:
      L *= (x-X[j])/(X[i]-X[j])
  return L

#cuando evaluamos Li(xj) cuando i=j
for i in range(len(X)):
  print(Lagrange(X[i],X,i))
```

Ahora si hacemos que $L_{i}(x_{j})$ cuando $i\neq j$

```python
#ahora si hacemos que Li(xj) cuando i!=j
for i in range(len(X)):
  if i<len(X)-1:
    print(Lagrange(X[i+1],X,i))
  if i==len(X)-1:
    print(Lagrange(X[0],X,i))
```

Esto nos muestra que $L_{i}(x) = δ_{ij}$ para cada $j ∈ {0, 1, ..., n}$

3. ¿Con que grado de exactitud podemos calcular $\sqrt{114}$ mediante la interpolacion de
Lagrange para la funcion $f(x) = \sqrt{x}$, si elegimos los puntos $x_{0} = 100, x_{1} = 121, x_{2} = 144$.


```python
X2=np.array([100,121,144])
Y2=np.array(np.sqrt(X2))
x=sym.Symbol('x')
def Interpolate(x,X,Y):

  Poly = 0
  for i in range(len(X)):
    Poly += Y[i]*Lagrange(x,X,i)
  return Poly

x1 = np.linspace(X2[0],X2[-1],50)
y1 = Interpolate(x1,X2,Y2)
plt.scatter(X2,Y2,color='r')
plt.scatter([114,114],[np.sqrt(114),Interpolate(x,X2,Y2).subs(x,114)])
plt.plot(x1,y1),plt.plot(x1,np.sqrt(x1))
plt.show()
```

```python
E=np.sqrt(114)-Interpolate(x,X2,Y2).subs(x,114)
E
```

Como $E=1.188*10^{-3}$, sabemos que la exactitud fue del $99.9\%$
